b. Imputation

```python
# 1st way of Imputation (Using simple python code)

data1 = pd.DataFrame(dataset)
print("Before Imputation\n")
print(data1.info())

for i in range(0,len(data1.axes[1])):
  if data1.iloc[:, i].isna().sum() > 0:
     data1.iloc[:, i]= data1.iloc[:, i].fillna( data1.iloc[:, i].mean())

print("\n\nAfter Imputation\n")
print(data1.info())
```

```python
# 2nd way of Imputation (Using python SimpleImputer)

import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

data1 = pd.DataFrame(dataset)
print("Before Imputation\n")
print(data1.info())

imputer = SimpleImputer(missing_values = np.nan,strategy = "mean")

for i in range(0,len(data1.axes[1])):
  if data1.iloc[:, i].isna().sum() > 0 and (data1.iloc[:, i].dtype == "int64" or data1.iloc[:, i].dtype == "float64"):
     # Fitting the data, function learns the stats
     imputer = imputer.fit(data1.iloc[:, i:i+1])
     # fit_transform() will execute those stats on the input ie. X[:, 1:3]
     data1.iloc[:, i] = imputer.transform(data1.iloc[:, i:i+1])

print("\n\nAfter Imputation\n")
print(data1.info())
```

# Correlation

Import the necessary libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.DataFrame(dataset)
data.drop(data.columns[[1, 11]], axis=1, inplace=True)
corr = data.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
# we compare the correlation between features and remove one of two features that have a correlation higher than 0.9

columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
  for j in range(i+1, corr.shape[0]):
    if corr.iloc[i,j] >= 0.9:
      if columns[j]:
         columns[j] = False
```

```python
selected_columns = data.columns[columns]
selected_columns.shape
```

```python
data = data[selected_columns]
print(data)
```
