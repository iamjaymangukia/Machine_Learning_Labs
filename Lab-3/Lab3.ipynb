# Exercise: 
Manually calculate output for the following cases and compare it with system’s output.
1. Will you play if the temperature is ‘Hot’ and weather is ‘overcast’?
2. Will you play if the temperature is ‘Mild’ and weather is ‘Sunny’?

Import necessary libraries.

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB, CategoricalNB
```

Prepare dataset.

```python
weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy','Rainy', 'Overcast',
'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy']

temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild',
'Cool','Mild','Mild','Mild','Hot','Mild']

play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes',
'Yes','Yes','Yes','Yes','No']
```

Digitize the data set using encoding

```python
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
weather_encoded=le.fit_transform(weather)
print("Weather:" ,weather_encoded)
```

```python
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)
print("Temp:",temp_encoded)
print("Play:",label)
```

Merge different features to prepare dataset

```python
#Combinig weather and temp into single listof tuples
features=tuple(zip(weather_encoded,temp_encoded))
print("Features:",features)
```

Train 'Naive Bayes Classifier’

```python
#Create a MultinomialNB Classifier
model=MultinomialNB()
# Train the model using the training sets
model.fit(features,label)

#Create a CategoricalNB Classifier
model1 = CategoricalNB()
# Train the model using the training sets
model1.fit(features,label)

#Create a GaussianNB Classifier
model2 = GaussianNB()
# Train the model using the training sets
model2.fit(features,label)
```

Predict Output for new data

```python
#Predict Output using MultinomialNB Classifier
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print("Predicted Value:", predicted)

#Predict Output using CategoricalNB Classifier
predicted= model1.predict([[0,2]]) # 0:Overcast, 2:Mild
print("Predicted Value:", predicted)

#Predict Output using GaussianNB Classifier
predicted= model2.predict([[0,2]]) # 0:Overcast, 2:Mild
print("Predicted Value:", predicted)
```

1. Will you play if the temperature is 'Hot' and weather is 'overcast'?

```python
predicted= model.predict([[0,1]]) # 0:Overcast, 1:Hot
print("Predicted Value:", predicted)

predicted= model1.predict([[0,1]]) # 0:Overcast, 1:Hot
print("Predicted Value:", predicted)

predicted= model2.predict([[0,1]]) # 0:Overcast, 1:Hot
print("Predicted Value:", predicted)
```

2. Will you play if the temperature is 'Mild' and weather is 'Sunny'?

```python
predicted= model.predict([[2,2]]) # 2:Sunny, 2:Mild
print("Predicted Value:", predicted)

predicted= model1.predict([[2,2]]) # 2:Sunny, 2:Mild
print("Predicted Value:", predicted)

predicted= model2.predict([[2,2]]) # 2:Sunny, 2:Mild
print("Predicted Value:", predicted)
```
